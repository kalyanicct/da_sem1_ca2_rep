```python
import pandas as pd
import requests
import json
```

```python
api_url = 'https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/THA25/JSON-stat/1.0/en'
```

```python
response = requests.get(api_url)
```

```python
if response.status_code == 200:
    json_data = response.json()
    print(json_data)
else:
    print(f"Error: {response.status_code}")
```
